```python
df_1['attack_point'] = df_1['goals'] + df_1['assist']
df_1
```

```python
df_1['player_name'] = df_1['first_name'] +' ' +  df_1['last_name']
df_1['player_name']
```

팀의 핵심 선수 찾기

```python
df_1['full_match'].sort_values()
df_1
```

```python
top_10_by_time = df_1.sort_values(by='Mt', ascending=False).head(10)
top_10_by_time

# 데이터 시각화
plt.figure(figsize=(12, 8))
sns.barplot(x = 'Mt', y = 'player_name', data = top_10_by_time)
plt.title('Top 10 Players by Minutes Played', fontsize=16)
plt.xlabel('Minutes Played')
plt.ylabel('Player name')

plt.savefig('../images/1_minutes_played.png')
plt.show()
```

```python
top_attackers = df_1[df_1['attack_point'] > 0].sort_values(by='attack_point', ascending= False).head(10)

plot_data = top_attackers[['player_name', 'goals', 'assist']].set_index('player_name')
plot_data.plot(kind = 'bar', stacked = True, figsize = (12, 8), colormap='viridis')
plt.title('Top 10 Offensive Contributors (Goals + Assists)', fontsize = 16)
plt.ylabel('Count')
plt.xticks(rotation = 45)

plt.savefig('../images/2_top_attacker.png')
plt.show()
```
